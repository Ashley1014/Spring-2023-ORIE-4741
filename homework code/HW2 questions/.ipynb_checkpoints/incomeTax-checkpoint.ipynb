```python
import pandas
```

```python
# load data
tax = pandas.read_csv("incomeTaxData.csv")
tax
```

You can see just from the first line of the dataset that the there are some rows with no returns associated with it. Let's delete those from the dataset.

Also note that all of the data is in thousands of dollars.

```python
# clean data
tax = tax.dropna(subset=['ReturnCount'])
tax = tax.sort_values(by=['Year'])

# create new columns
tax['avg_tax'] = tax['TotalTaxLiability']/tax['ReturnCount']
```

## a)


Plot the number of returns in Tompkins County from each income class bracket over time on a single graph.（Ignore the rows with the class of 'Total'. Should span 1999-2013. Some of the county names may have leading/trailing whitespaces.）

```python
import numpy as np
import matplotlib.pyplot as plt

tompkins = tax.loc[tax['County'].str.strip() == "Tompkins"]
tompkins = tompkins.loc[tompkins['IncomeClass'] != "Total"]
income_classes = tompkins.IncomeClass.unique()
returns_by_year = {}
NOY = 2013 - 1999 + 1
for ic in income_classes:
    returns_by_year[ic] = [0] * NOY

for idx, row in tompkins.iterrows():
    income_class = row['IncomeClass']
    year = row['Year']
    year_idx = year - 1999
    count = row['ReturnCount']
    returns_by_year[income_class][year_idx] = count

fig = plt.figure(figsize = (20, 10))
for ic in returns_by_year:
    years = np.arange(1999,2014)
    counts = np.array(returns_by_year[ic])
    plt.plot(years, counts, label=ic)

plt.xlabel("Year")
plt.ylabel("Tax Returns")
plt.title("Tax returns in Tompkins County from 1999 to 2013 by income classes")
plt.legend(loc="upper right")
plt.show()
```

Plot the average income tax per return in Tompkins County (disregarding income tax).（You could add up the all income classes (except 'Total') to calculate the weighted average tax OR just plot the rows of 'Total'. The sum might differ from the number in 'total'; either answer is ok. Messy data!)

```python
tompkins = tax.loc[tax['County'].str.strip() == "Tompkins"]
tompkins = tompkins.loc[tompkins['IncomeClass'] == "Total"]

avg_per_return = [0] * NOY
for idx, row in tompkins.iterrows():
    year = row['Year']
    year_idx = year - 1999
    income_tax = row['avg_tax']
    avg_per_return[year_idx] = income_tax

fig = plt.figure(figsize = (20, 10))
years = np.arange(1999,2014)
avg_per_return = np.array(avg_per_return)
plt.plot(years, avg_per_return)
plt.xlabel("Year")
plt.ylabel("avg_tax per return")
plt.title("Average income tax per return in Tompkins County from 1999 to 2013")
plt.show()
```

What kind of plot did you choose to make? Why?
I choose to make line graph, because it can clearly describe the trend of data over time.

## b)

Continuing to look only at Tompkins County, fit a model that predicts avg_tax using the year number. Do this with least squares.

```python
Y = np.array(tompkins['avg_tax'].values, 'float')
x = np.arange(1999,2014).reshape(-1, 1)
w = np.linalg.lstsq(x, Y)[0]
print(w)
```

Transform the input to include an offset term in the model

```python
n = Y.shape[0]
X = np.column_stack([x, np.ones(n)])
wb = np.linalg.lstsq(X, Y)[0]
print(wb)
```

## c)

Now we will change the feature space. Fit a model using the year and the avg_tax from the previous year (and offset).

```python
from sklearn.linear_model import LinearRegression

prevs = [0] + [tompkins.loc[tompkins['Year'] == y]['avg_tax'].values[0] for y in years[:-1]]
prevs = np.array(prevs)
X = np.column_stack([x, prevs, np.ones(n)])
reg = LinearRegression().fit(X, Y)
wc = np.array([reg.coef_[0], reg.coef_[1], reg.intercept_])
print(wc)
```

Interpret the coefficients of this. What do they mean?

The first coefficient means that when other conditions held, the avg_tax increases by 5.05890753e-02 unit by year increase of one year. The second coefficient means that when the other conditions held, the avg_tax of next year increases by 2.17512032e-02 unit compared to last year's avg_tax. The third coefficient means that the regression line crossing the line of x = 0 at y = -9.93962089e+01.

Plot the real average tax return per year against your model's prediction. Discuss how you think your model fits.

The overall trend of my model seems right, but there are some fluctuating points that are not correctly captured by my model.

```python
y_hat = reg.predict(X)
fig = plt.figure(figsize = (20, 10))
years = np.arange(1999,2014)
plt.scatter(years, y_hat, label="predicted_y")
plt.scatter(years, Y, label="actual_y")
plt.xlabel("Year")
plt.ylabel("avg_tax")
plt.title("Predicted and actual avg_tax in Tompkins County from 1999 to 2013")
plt.legend(loc="upper right")
plt.show()
```

## d)

Add two new features to your model. This could be a column of the data, a transformation of a column, or a new column formed from another. The only requirement is that the prediction in year $t$ be based on data avilable by the end of year $t-1$

State the feature space for your new model and why you think those features predict the data well. Fit this model and interpret the coefficients.

The new feature space for my model would be the year, the avg_tax from the previous year, the total income, and the total taxable income. I think these features will predict the data well because tax will be largely dependent on income.

The first coefficient means that when other conditions held, the avg_tax decreases by -9.43228670e-03 unit by year increase of one year. The second coefficient means that when the other conditions held, the avg_tax of next year increases by 1.38951827e-02 unit compared to last year’s avg_tax. The third coefficient means that when the other conditions held, the avg_tax increases by 1.74537316e-07 unit by one unit increase of total income of the county. The fourth coefficient means that when the other conditions held, the avg_tax increases by 1.03362471e-06 unit by one unit increase of total taxable income of the county. The fifth coefficient means that the regression line crossing the line of x = 0 at y = 1.92505774e+01.

```python
total_income = np.array(tompkins['TotalIncome']).reshape(-1, 1)
taxable_income = np.array(tompkins['TotalTaxableIncome']).reshape(-1, 1)
X_new = np.column_stack([x, prevs, total_income, taxable_income, np.ones(n)])
reg = LinearRegression().fit(X_new, Y)
wd = np.array([reg.coef_[0], reg.coef_[1], reg.coef_[2], reg.coef_[3], reg.intercept_])
print(wd)
```

## e)

Compare the coefficients from part c and d. Does the coefficient of avg_tax differ in the two models? If so, how do you interpret this difference?

We can see that in the new model, the coefficient associated with year is no longer positive, and the coefficient associated with previous year's avg_tax also decrease, that is because the total income of the county and the total taxable income of the county have positive effect on avg_tax, and these were included in the original effects observed on year or prev_avg_tax without being explained correctly.

## f)

(see homework handout) Now we want to see how this model performs in other counties. Apply the model you fit in part c, with coefficients $w^c$, to data from the other counties and compute the error of this model on each data set.

Plot a histogram of the mean errors the model makes.

```python
from sklearn.metrics import mean_absolute_error
def get_avg_tax(county):
    county = tax.loc[tax['County'].str.strip() == county]
    county = county.loc[county['IncomeClass'] == "Total"]
    avg_per_return = [0] * NOY
    for idx, row in county.iterrows():
        year = row['Year']
        year_idx = year - 1999
        income_tax = row['avg_tax']
        avg_per_return[year_idx] = income_tax
    return avg_per_return
counties = tax['County'].str.strip().unique()
# remove fake counties
filtered_counties = filter(lambda county: county not in ["NYS Unclassified +", "Residence Unknown ++", "Grand Total", "Full-Year Resident"], counties)

errors = {}
for county in filtered_counties:
    actual_y = np.array(get_avg_tax(county))
    reg = LinearRegression().fit(X, actual_y)
    predicted_y = reg.predict(X)
    error = mean_absolute_error(actual_y, predicted_y)
    errors[county] = error

plt.hist(errors.values())
plt.show()
```

Are there major outliers?

There are some major outliers, with an error of around 0.4-0.7.

Again, are there major outliers? How do the coefficients compare across counties?

There still are some major outliers, with mean square errors ranging from 0.3-0.6. The coefficients differ significantly in different counties.
